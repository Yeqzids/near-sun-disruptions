# Orbits of $q<0.2$ AU and $T_\mathrm{J}>2$ objects, including SOHO objects

```python
%matplotlib inline 

import matplotlib.pyplot as plt, numpy as np
from matplotlib.ticker import FormatStrFormatter

plt.rc('font', family='serif')
plt.rc('text', usetex=True)
```

```python
import numpy as np

d = np.genfromtxt('../data/soho.dat', skip_header=1, delimiter=',').T
```

```python
def sq(x, a):
    return a*x**2
```

```python
fig = plt.figure(figsize=(3.25, 2.75), dpi=150)

ax = fig.add_subplot(111)

ax.hist([d[6][:-6], d[6][-5:]], 15, histtype='bar', stacked='True', label=['Asteroid', 'SOHO object'])

x = np.arange(0.05, 0.2, 0.01)
ax.plot(x, sq(x, 300), 'k--', linewidth=1, label='Crude $q^2$ fit')

ax.legend(fontsize='x-small')
ax.set_xlabel('$q$ (au)')
ax.set_ylabel('Cumulative number of objects')
ax.set_yticks(np.arange(0, 20, step=5))

fig.tight_layout()
fig.savefig('soho-orb.pdf')
plt.show()
```
